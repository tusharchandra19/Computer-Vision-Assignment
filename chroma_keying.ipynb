```python
import cv2
import numpy as np
 
# Create a VideoCapture object
cap = cv2.VideoCapture("Tushar.mp4")
 
# Check if camera opened successfully
if (cap.isOpened() == False): 
    print("Unable to read camera feed")
 
# We convert the resolutions from float to integer.
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
i=1
# Define the codec and create VideoWriter object.The output is stored in 'outpy.avi' file. 
while(True):
    ret, frame = cap.read()
    
    if ret == True: 
        # Write the frame into the file 'output.avi
        # Display the resulting frame    
        cv2.imshow('frame',frame)
        cv2.imwrite('new_img'+str((i//100)%10)+str((i//10)%10)+str(i%10)+'.jpg',frame)
        # Press Q on keyboard to stop recording
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    # Break the loop
    else:
        break 
    i+=1
 
# When everything done, release the video capture and video write objects
cap.release()
 
# Closes all the frames
cv2.destroyAllWindows() 
```

```python
import glob

img_array = []
for filename in glob.glob("C:/Users/Tushar/Computer Vision/*.jpg"):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)


output = cv2.VideoWriter('Tushar1.avi',cv2.VideoWriter_fourcc(*'DIVX'), 30, size)
 
for i in range(len(img_array)):
    output.write(img_array[i])
output.release()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
i=1
while i<101:
    image = cv2.imread('new_img'+str(i)+'.jpg')
    image_copy = np.copy(image)
    lower_green = np.array([145, 145, 145])    
    upper_green = np.array([255, 255, 255])
    mask = cv2.inRange(image_copy, lower_green, upper_green)
    masked_image = np.copy(image_copy)

    masked_image[mask != 0] = [0, 0, 0]
    background_image = cv2.imread('eifel1'+'.jpg')
    # Cropping it to the right size
    crop_background = background_image[0:352, 0:640]
    # Mask the cropped background so that the pizza area is blocked
    
    crop_background[mask == 0] = [0, 0, 0]
    final_image = crop_background + masked_image
    
    cv2.imwrite('final_image'+str((i//100)%10)+str((i//10)%10)+str(i%10)+'.jpg',final_image)
    i+=1
    

    
```
